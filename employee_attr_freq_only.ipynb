```python
#importation des packages
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

employee_attrition = pd.read_csv(r"C:\Disque Perso 2023-10-20\DATA\Employee attrition\employee_attrition_dataset_10000.csv")

age_range = [20,25,30,35,40,45,50,55,60]
age_cat = ['[20-25[','[25-30[','[30-35[','[35-40[','[40-45[','[45-50[','[50-55[','[55-60[']
employee_attrition['Age_category'] = pd.cut(employee_attrition['Age'], bins=age_range, labels = age_cat, right=False,include_lowest=True)

attrition_filter = employee_attrition['Attrition']=='Yes'
attrition_leave = employee_attrition[attrition_filter]
attrition_leave
```

```python
sns.countplot(data=attrition_leave, x="Age_category")
plt.title("attrition frequency per age")
plt.show()

sns.catplot(data=attrition_leave, x="Overtime", kind='count')
plt.title("attrition frequency per overtimme")
plt.show()
```

```python
sns.catplot(data=attrition_leave, x="Gender", kind='count')
plt.title("attrition frequency per Gender")
plt.show()

sns.catplot(data=attrition_leave, x="Marital_Status", kind='count')
plt.title("attrition frequency per Marital_Status")
plt.show()

sns.catplot(data=attrition_leave, x="Job_Role", kind='count')
plt.title("attrition frequency per Job_Role")
plt.show()

sns.catplot(data=attrition_leave, x="Job_Level", kind='count')
plt.title("attrition frequency per Job_Level")
plt.show()

sns.catplot(data=attrition_leave, x="Work_Life_Balance", kind='count')
plt.title("attrition frequency per Work_Life_Balance")
plt.show()
```

```python
import pandas as pd
employee_attrition = pd.read_csv(r"C:\Disque Perso 2023-10-20\DATA\Employee attrition\employee_attrition_dataset_10000.csv")
```

```python
#Data exploration, validation & cleaning
# Checks all data types
employee_attrition.info()
# check null values
employee_attrition.isna().sum()
```

```python
# Check categorical data
Gender_value=list(employee_attrition["Gender"].unique())
Marital_Status=list(employee_attrition["Marital_Status"].unique())
Department=list(employee_attrition["Department"].unique())
Job_Roles=list(employee_attrition["Job_Role"].unique())
Overtime=list(employee_attrition["Overtime"].unique())
Attrition=list(employee_attrition["Attrition"].unique())
print(f"gender values are {Gender_value}")
print(f"Marital_Status values are {Marital_Status}")
print(f"Department values are {Department}")
print(f"Job_Roles values are {Job_Roles}")
print(f"Overtimes values are {Overtime}")
print(f"Attrition values are {Attrition}")
```

DATA CLEANING CONCLUSION
Data is already clean, there is no null values. All categorical values respect its due values.
All data types are correct.

```python
#DATA EXPLORATION
#Explore num√©rical data
employee_attrition.describe().round(2)
```

```python
print(employee_attrition.groupby('Attrition')['Age'].mean())
```

```python
#Check relation between attrition and other criterias
import matplotlib.pyplot as plt
import seaborn as sns
palette_attrition = {'Yes':'purple','No':'pink'}

sns.catplot(data=employee_attrition["Attrition"], kind="count", color='pink')
plt.title("attrition frequency")
plt.show()

attrition_freq = employee_attrition["Attrition"].value_counts(normalize=True).round(2)
attrition_freq



```

```python
#grouping employees into age categories
employee_attrition["Age"].describe()
```

```python
age_range = [20,25,30,35,40,45,50,55,60]
age_cat = ['[20-25[','[25-30[','[30-35[','[35-40[','[40-45[','[45-50[','[50-55[','[55-60[']
employee_attrition['Age_category'] = pd.cut(employee_attrition['Age'], bins=age_range, labels = age_cat, right=False,include_lowest=True)
employee_attrition['Age_category']
palette_attrition = {'Yes':'purple','No':'pink'}

sns.countplot(data=employee_attrition, x="Age_category", hue="Attrition", dodge=False, palette = palette_attrition)
plt.title("attrition frequency per age")
plt.show()


sns.countplot(data=employee_attrition, x="Job_Level", hue="Attrition", dodge=False, palette = palette_attrition)
plt.title("attrition frequency per job level")
plt.show()

sns.catplot(data=employee_attrition, x="Overtime", hue="Attrition", kind='count', dodge=False, palette = palette_attrition)
plt.title("attrition frequency per overtimme")
plt.show()

```

```python
print(employee_attrition[employee_attrition['Attrition']=='Yes']['Age_category'].value_counts(normalize=True).round(2))
print(employee_attrition[employee_attrition['Attrition']=='Yes']['Job_Level'].value_counts(normalize=True).round(2))




```
